```python
df["Probability"]=result
```

```python
chance_of_admit_threshold=0.69
def chance(var):
    a = []


    for i in df[var]:
        if i >=chance_of_admit_threshold:
            i="YES"
        else:
            i="NO"
        a.append(i)
    return a
df["chance_of_admit"]=chance("Probability")
    
```
